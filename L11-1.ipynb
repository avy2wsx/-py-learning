```python
fig = plt.figure()
ax_1 = fig.add_subplot(1,3,1)
ax_1.set_title("axes1")

ax_2 = fig.add_subplot(1,3,2)
ax_2.set_title("axes2")

ax_3 = fig.add_subplot(1,3,3)
ax_3.set_title("axes3")
```

```python
student1 = {'chinese':78,'english':80,'history':75,'discovery':85}
student2 = {'chinese':88,'english':75,'history':84,'discovery':76}

subjects = list(student1.keys())
value1 = list(student1.values())
value2 = list(student2.values())

figure = plt.figure(figsize=(8.0,5.0),dpi=250.0,facecolor="#cccccc")
ax =figure.add_subplot()
ax.plot(subjects,value1,'b-.o',lw=3)
ax.plot(subjects,value2,'m--^',lw=5)

plt.show()
```
